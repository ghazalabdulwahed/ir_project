```python
import pandas as pd
import numpy as np

from sentence_transformers import SentenceTransformer
```

```python
df = pd.read_csv("quora_cleaned_sample.csv")

print(df.shape)
```

```python
sample_docs = df["cleaned_text"].astype(str).tolist()[:50]

print("Number of documents:", len(sample_docs))
print(sample_docs[:3])
```

```python
embeddings = model.encode(
    sample_docs,
    show_progress_bar=True
)

print(embeddings.shape)
```

```python
np.save("sample_embeddings.npy", embeddings)

print("Saved Successfully")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
query = "invest in stock market"
```

```python
query_embedding = model.encode([query])

print(query_embedding.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

scores = cosine_similarity(
    query_embedding,
    embeddings
)[0]

print(scores[:5])
```

```python
top_indices = scores.argsort()[::-1]

print(top_indices[:5])
```

```python
for idx in top_indices[:5]:

    print("Similarity Score:", round(scores[idx], 4))

    print("Document:")
    print(sample_docs[idx])

    print("-" * 50)
```

```python
all_docs = df["cleaned_text"].astype(str).tolist()

print("Number of documents:", len(all_docs))
```

```python
all_embeddings = model.encode(
    all_docs,
    batch_size=32,
    show_progress_bar=True
)
```
