```python
import ir_datasets

def check_project_conditions(dataset_name):
    print(f"🔍 جاري فحص مجموعة البيانات: ({dataset_name})")
    
    # تحميل المجموعة
    dataset = ir_datasets.load(dataset_name)
    is_valid = True
    
    # 1. فحص شرط عدد الوثائق (أكبر من 200 ألف)
    if dataset.has_docs():
        doc_count = dataset.docs_count()
        print(f"  📄 عدد الوثائق الكلي (Documents): {doc_count:,}")
        if doc_count > 200000:
            print("  ✅ شرط الحجم: محقق (أكبر من 200,000 وثيقة).")
        else:
            print("  ❌ شرط الحجم: غير محقق (أقل من 200,000 وثيقة)!")
            is_valid = False
    else:
        print("  ❌ خطأ: المجموعة لا تحتوي على وثائق!")
        is_valid = False

    # 2. فحص شرط وجود استعلامات (Queries)
    if dataset.has_queries():
        print("  ❓ شرط الاستعلامات (Queries): محقق وجاهز ✅")
        # طباعة عينة للتأكد من القراءة الصحيحة
        for q in dataset.queries_iter():
            print(f"     [عينة استعلام] ID: {q.query_id} | النص: {q.text[:50]}...")
            break
    else:
        print("  ❌ شرط الاستعلامات (Queries): غير محقق!")
        is_valid = False

    # 3. فحص شرط وجود تقييمات الملائمة (Qrels)
    if dataset.has_qrels():
        print("  🎯 شرط التقييمات (Qrels): محقق وجاهز ✅")
        # طباعة عينة للتأكد
        for qrel in dataset.qrels_iter():
            print(f"     [عينة تقييم ملائمة] Query ID: {qrel.query_id} -> مرتبط بالوثيقة Doc ID: {qrel.doc_id}")
            break
    else:
        print("  ❌ شرط التقييمات (Qrels): غير محقق!")
        is_valid = False
        
    if is_valid:
        print(f"\n🎉 النتيجة لـ {dataset_name}: المجموعة مطابقة لشروط المشروع 100% ومحملة بشكل صحيح!")
    else:
        print(f"\n⚠️ النتيجة لـ {dataset_name}: هناك خلل في الشروط!")
    print("\n" + "="*50 + "\n")

# تشغيل الفحص على المجموعتين اللتين قمتِ بتحميلهما
check_project_conditions("beir/quora/test")
check_project_conditions("msmarco-passage/train")
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# مصفوفة لتخزين البيانات بعد التنظيف
corpus_sample = []

print("⏳ جاري تنظيف ومعالجة عينة البيانات للجروب...")

# dataset1 هي مجموعة Quora التي تم تحميلها في الخلية السابقة
for i, doc in enumerate(dataset1.docs_iter()):
    
    # 1. تحويل النص لأحرف صغيرة وتقطيعه إلى كلمات
    words = word_tokenize(doc.text.lower())
    
    # 2. حذف كلمات التوقف (Stopwords) والتأكد من أنها كلمات وليست رموزاً
    filtered_words = [w for w in words if w not in stopwords.words('english') and w.isalnum()]
    
    # 3. إعادة دمج الكلمات المنظفة في نص واحد
    cleaned_text = " ".join(filtered_words)
    
    # حفظ البيانات في القائمة
    corpus_sample.append({
        'doc_id': doc.doc_id,
        'original_text': doc.text,
        'cleaned_text': cleaned_text
    })
    
    # التوقف عند 10,000 وثيقة ليكون الملف خفيفاً وسريعاً على غيت هاب
    if i >= 10000:
        break

# تحويل القائمة إلى جدول وحفظه كملف CSV
df = pd.DataFrame(corpus_sample)
df.to_csv("quora_cleaned_sample.csv", index=False)

print("✅ نجاح كامل! تم إنشاء ملف 'quora_cleaned_sample.csv' في نفس مجلد مشروعك.")
print("هذا هو الملف الخفيف الذي سترفعونه على GitHub لتشتغلوا عليه جميعاً!")
```
