```python
df = df.dropna(subset=['cleaned_text'])
df['cleaned_text'] = df['cleaned_text'].fillna('')
```

```python
from collections import defaultdict

inverted_index = defaultdict(set)

for _, row in df.iterrows():
    doc_id = row['doc_id']

    text = str(row['cleaned_text'])
    words = text.split()

    for word in words:
        inverted_index[word].add(doc_id)

print(" تم بناء الفهرس المعكوس بنجاح")
print("عدد الكلمات المميزة:", len(inverted_index))
```

```python
import math
from collections import Counter
N = df.shape[0]
tf_docs = {} 
for _, row in df.iterrows():
    doc_id = row['doc_id']
    words = row['cleaned_text'].split()
    word_counts = Counter(words)
    total_words = len(words)
    tf_docs[doc_id] = {word: count/total_words for word, count in word_counts.items()}

print(" تم حساب TF لكل وثيقة")
```

```python
df_counts = {word: len(doc_ids) for word, doc_ids in inverted_index.items()}
idf = {word: math.log(N / df_count) for word, df_count in df_counts.items()}

print(" تم حساب IDF لكل كلمة")
```

```python
tfidf_docs = {}

for doc_id, tf_dict in tf_docs.items():
    tfidf_docs[doc_id] = {word: tf * idf[word] for word, tf in tf_dict.items()}

print(" تم حساب TF-IDF لكل وثيقة")
```

```python
first_doc_id = df['doc_id'].iloc[0]
print(f"TF-IDF للوثيقة {first_doc_id}:")
for word, score in list(tfidf_docs[first_doc_id].items())[:5]:
    print(f"{word}: {score:.4f}")
```

```python
doc_lengths = {}

for _, row in df.iterrows():
    doc_id = row['doc_id']
    words = row['cleaned_text'].split()
    doc_lengths[doc_id] = len(words)

avg_doc_length = sum(doc_lengths.values()) / len(doc_lengths)

print("متوسط طول الوثائق:")
print(avg_doc_length)
```

```python
df_counts = {}

for word, docs in inverted_index.items():
    df_counts[word] = len(docs)

print("عدد الكلمات:", len(df_counts))
```

```python
import math

N = len(df)

bm25_idf = {}

for term, df_term in df_counts.items():
    bm25_idf[term] = math.log(
        ((N - df_term + 0.5) / (df_term + 0.5)) + 1
    )

print("تم حساب BM25 IDF")
```

```python
from collections import Counter

def bm25_search_fast(query, k1=1.5, b=0.75):
    scores = {}
    query_terms = query.lower().split()
    candidate_docs = set()
    for term in query_terms:
        if term in inverted_index:
            candidate_docs.update(inverted_index[term])
            

    for doc_id in candidate_docs:

        doc_row = df[df['doc_id'] == doc_id].iloc[0]
        words = str(doc_row['cleaned_text']).split()
        
        doc_len = len(words)
        term_freqs = Counter(words)
        score = 0
        
        for term in query_terms:
            if term not in term_freqs:
                continue
                
            tf = term_freqs[term]
            idf = bm25_idf.get(term, 0)
            
            numerator = tf * (k1 + 1)
            denominator = tf + k1 * (1 - b + b * (doc_len / avg_doc_length))
            
            score += idf * (numerator / denominator)
            
        scores[doc_id] = score

    ranked_docs = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    return ranked_docs
```

```python
results = bm25_search(
    "machine learning",
    k1=1.5,
    b=0.75
)

print("أفضل 10 نتائج:\n")

for doc_id, score in results[:10]:
    print(doc_id, round(score, 4))
```
