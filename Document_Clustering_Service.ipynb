```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("quora_cleaned_sample.csv")

print(df.shape)
df.head()
```

```python
import os

print(os.listdir())
```

```python
import numpy as np

embeddings = np.load("quora_embeddings.npy")

print(embeddings.shape)
```

```python
from sklearn.cluster import KMeans

k = 10

kmeans = KMeans(
    n_clusters=k,
    random_state=42,
    n_init=10
)

clusters = kmeans.fit_predict(embeddings)

print(clusters[:20])
```

```python
df["cluster"] = clusters

df.head()
```

```python
for i in range(10):
    print(f"\n===== Cluster {i} =====")

    docs = df[df["cluster"] == i]["cleaned_text"].head(5)

    for doc in docs:
        print("-", doc)
```

```python
print(df["cluster"].value_counts())
```

```python
for i in range(10):
    print(f"\n===== Cluster {i} =====")

    sample = df[df["cluster"] == i]["cleaned_text"].head(5)

    for text in sample:
        print("-", text)
```

```python
df.to_csv("clustered_documents.csv", index=False)
```
